```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,accuracy_score,confusion_matrix,classification_report
```

```python
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, pred)
r2 = r2_score(y_test, pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python


from sklearn.metrics import mean_squared_error

y_pred_train = rf.predict(x_train)
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))

print("Training RMSE:", rmse_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
model=SVR()
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

grid_search = GridSearchCV(model, param_grid, cv=5, n_jobs=-1, verbose=3)
grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_
```

```python

model = Sequential()
model.add(Dense(128, input_dim=X.shape[1], activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    x_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

```python
nn_pred = model.predict(x_test).flatten()
def evaluate_model(y_test, predictions, model_name):
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    print(f"Model: {model_name}")
    print(f"Mean Squared Error: {mse}")
    print(f"Root Mean Squared Error: {rmse}")
    print(f"Mean Absolute Error: {mae}")
    print(f"R-squared: {r2}\n")
    return {'Model': model_name, 'MSE': mse, 'RMSE': rmse, 'MAE': mae, 'R2': r2}

nn_metrics = evaluate_model(y_test, nn_pred, "Neural Network (TensorFlow)")
```

```python
rf_pred=rf.predict(x_test)
rf_metrics = evaluate_model(y_test, rf_pred, "Random Forest Regressor")
svr_pred = grid_search.predict(x_test)
svr_metrics = evaluate_model(y_test, svr_pred, "Support Vector Regressor (SVR)")




results = pd.DataFrame([rf_metrics, svr_metrics, nn_metrics])
print("Model Performance Comparison with Neural Network:\n", results)


results.set_index('Model', inplace=True)
results[['MSE', 'RMSE', 'MAE']].plot(kind='bar', figsize=(10, 6), title="Error Metrics Comparison Across Models (Including NN)")
plt.ylabel("Error")
plt.show()
results['R2'].plot(kind='bar', figsize=(6, 4), color='purple', title="R2 Score Comparison Across Models (Including NN)")
plt.ylabel("R2 Score")
plt.show()
```
